```python
df=df.drop(columns = ['BuildingArea'])
print(df)
```

```python
df=df.drop_duplicates()
```

```python
df['CouncilArea']=df['CouncilArea'].fillna(df['CouncilArea'].mode()[0])
```

```python
df
```

```python
df=df.dropna(subset=["YearBuilt"])
df
```

```python
(df.isnull().sum()/len(df))*100
```

```python

df.loc[:,'Car']=df['Car'].fillna(df['Car'].mean())
df
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
df
```

```python
df.isnull().sum()
```

```python
df
```

```python
q1= df['Price'].quantile(0.25)
q3= df['Price'].quantile(0.75)
iqr = q3 - q1 
lower = q1 - 1.5*iqr
upper = q3 + 1.5*iqr

df= df[(df['Price']>=lower) & (df['Price']<=upper)]
df
```

```python
q1= df['Landsize'].quantile(0.25)
q3= df['Landsize'].quantile(0.75)
iqr = q3 - q1 
lower = q1 - 1.5*iqr
upper = q3 + 1.5*iqr

df= df[(df['Landsize']>=lower) & (df['Landsize']<=upper)]
df
```

```python
df
```

```python
df.to_excel("cleaned_data.xlsx",index=False)
```

```python
sns.histplot(df['Price'], bins=30)
plt.title("Distribution of House Prices")
plt.xlabel("Price")
plt.ylabel("Number of Houses")

plt.show()
```

```python

df.sort_values(by='Price')['Price'].head(10).plot(kind="bar",color='orange')
plt.title("Top 10 Most Expensive Suburbs")
plt.xlabel("Suburb")
plt.ylabel("Average Price")

plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=df,
    x='Landsize',
    y='Price'
)

plt.title("Land Size vs Property Price")

plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    data=df,
    x='Rooms',
    y='Price',
    color='red'
)

plt.title("Rooms vs Price")

plt.show()
```

```python
df.groupby('Rooms')['Price'].mean().plot(
    kind='line',
    marker='o',
)

plt.title("Average House Price by Number of Rooms")
plt.xlabel("Number of Rooms")
plt.ylabel("Average Price")

plt.show()
```
